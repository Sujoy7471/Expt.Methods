```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit

n = 1.50

# Load the Excel file
file_path = 'Fresnel_Data.xlsx'
df = pd.read_excel(file_path)

# For Perpendicular Polarization
xper = df.iloc[1:14, 0].tolist()
yper = df.iloc[1:14, 3].tolist()

plt.plot(xper, yper, linestyle = 'none', marker = "o", linewidth = 2, color = 'blue', markerfacecolor = 'red', label = "Experimental" )

u = np.linspace(0,90,1000)
v = np.zeros(len(u))
ur = u * np.pi / 180
def sqsin(n,ur):
    return np.sqrt(n*n - np.square(np.sin(ur)))

v = -(sqsin(n,ur) - np.cos(ur))**2/(n**2-1)

plt.plot(u,abs(v), linestyle = '-',color = 'blue', label = "Theoritical" )
plt.xlabel("Angles in Degree")
plt.ylabel("Reflection Coefficient")
plt.title("Reflection Coefficient vs Angles for Perpendicular Polarization")
plt.legend()
plt.grid()
plt.show()



# For Parallel Polarization
xpar = df.iloc[2:20, 5].tolist()
ypar = df.iloc[2:20, 8].tolist()

plt.plot(xpar, ypar, linestyle = 'none', marker = "o", linewidth = 2, color = 'blue', markerfacecolor = 'red', label = "Experimental" )

v = (n*n*np.cos(ur) - np.sqrt(n*n - np.square(np.sin(ur))))/(n*n*np.cos(ur) + np.sqrt(n*n - np.square(np.sin(ur))))

plt.plot(u,abs(v), linestyle = '-',color = 'blue', label = "Theoritical" )
plt.xlabel("Angles in Degree")
plt.ylabel("Reflection Coefficient")
plt.title("Reflection Coefficient vs Angles for Parallel Polarization")
plt.legend()
plt.grid()
plt.show()


```
