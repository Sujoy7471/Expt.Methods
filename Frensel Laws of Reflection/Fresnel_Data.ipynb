```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


n = 1.50

# Load the Excel file
file_path = 'Fresnel_Data.xlsx'
df = pd.read_excel(file_path)


# For Perpendicular Polarization
xper = df.iloc[1:14, 0].tolist()
yper = df.iloc[1:14, 4].tolist()

plt.plot(xper, yper, linestyle = 'none', marker = "o", linewidth = 2, color = 'blue', markerfacecolor = 'red', label = "Experimental" )

u = np.linspace(0,90,1000)
v = np.zeros(len(u))
ur = u * np.pi / 180
def sqsin(n,ur):
    return np.sqrt(n*n - np.square(np.sin(ur)))

v1 = -(sqsin(n,ur) - np.cos(ur))**2/(n**2-1)

plt.plot(u,abs(v1), linestyle = '-',color = 'blue', label = "Theoritical" )
plt.xlabel("Angles in Degree")
plt.ylabel("Reflection Coefficient")
plt.title("Reflection Coefficient vs Angles for Perpendicular Polarization")
plt.legend()
plt.grid()
plt.show()



# For Parallel Polarization
xpar = df.iloc[2:20, 6].tolist()
ypar = df.iloc[2:20, 10].tolist()

plt.plot(xpar, ypar, linestyle = 'none', marker = "o", linewidth = 2, color = 'blue', markerfacecolor = 'red', label = "Experimental" )

v2 = (n*n*np.cos(ur) - np.sqrt(n*n - np.square(np.sin(ur))))/(n*n*np.cos(ur) + np.sqrt(n*n - np.square(np.sin(ur))))

plt.plot(u,abs(v2), linestyle = '-',color = 'blue', label = "Theoritical" )
plt.xlabel("Angles in Degree")
plt.ylabel("Reflection Coefficient")
plt.title("Reflection Coefficient vs Angles for Parallel Polarization")
plt.legend()
plt.grid()
plt.show()


##Plot of alpha
x_alpha = df.iloc[1:13, 12].tolist()
y_alpha = df.iloc[1:13, 15].tolist()

plt.scatter(x_alpha, y_alpha, label = "Experimental" )

alpha = np.arctan(abs(v1)/abs(v2))*180/np.pi - 45

plt.plot(u,alpha, linestyle = '-',color = 'blue', label = "Theoritical" )
plt.xlabel("Angle of Incidence in Degree")
plt.ylabel("Change in angle of Polarization in Degree")
plt.title("Angle of incidence vs change in Angle of Polarization")
plt.legend()
plt.grid()
plt.show()

```
